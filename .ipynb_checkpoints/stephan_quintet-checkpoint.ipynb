## Read in data

```python
datadir='/Users/cosalyk/DATA/JWST/NGC7319/'
names=["jw02732-c1001_t004_miri_ch1-short","jw02732-c1001_t004_miri_ch1-medium","jw02732-c1001_t004_miri_ch1-long"]
keylist=['1A','1B','1C','2A','2B','2C','3A','3B','3C','4A','4B','4C']
channel_list=['ch1','ch2','ch3','ch4']
setting_list=['short','medium','long']

data_dict={}

#Loop through fits files and create a dictionary
k=0
for i,mychannel in enumerate(channel_list):
    for j,mysetting in enumerate(setting_list):
        filename=datadir+"jw02732-c1001_t004_miri_"+mychannel+"-"+mysetting+'/'+"jw02732-c1001_t004_miri_"+mychannel+"-"+mysetting+'_x1d.fits'
        f = fits.open(filename)  
        specdata = f[1].data 
        f.close() 
        wave=specdata['WAVELENGTH']
        flux=specdata['FLUX']
        error=specdata['FLUX_ERROR']
        mykey=keylist[k]
        spec_dict={'wave':wave,'flux':flux,'error':error}
        data_dict[mykey]=spec_dict
        k=k+1
```

```python
filename=datadir+"jw02732-c1001_t004_miri_ch1-longmediumshort-/jw02732-c1001_t004_miri_ch1-longmediumshort-_x1d.fits"

f = fits.open(filename)  
# The spectrum is in the second HDU of this file.
specdata = f[1].data 
f.close() 
#print(specdata.columns)
wave_comb=specdata['WAVELENGTH']
flux_comb=specdata['FLUX']
error_comb=specdata['FLUX_ERROR']
```
